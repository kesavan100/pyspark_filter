1. Core DataFrame Transformations (Daily Use — 80% of Real Work)

| Transformation         | Purpose                        | Example                            |
| ---------------------- | ------------------------------ | ---------------------------------- |
| `select()`             | Choose columns                 | `df.select("name","salary")`       |
| `withColumn()`         | Add / modify column            | `df.withColumn("x", col("a")+1)`   |
| `withColumnRenamed()`  | Rename column                  | `df.withColumnRenamed("a","A")`    |
| `drop()`               | Remove column                  | `df.drop("dept")`                  |
| `filter()` / `where()` | Filter rows                    | `df.filter(col("age")>30)`         |
| `orderBy()` / `sort()` | Sort rows                      | `df.orderBy(col("salary").desc())` |
| `distinct()`           | Remove duplicate rows          | `df.distinct()`                    |
| `groupBy()` + `agg()`  | Aggregation                    | `df.groupBy("dept").avg("salary")` |
| `join()`               | Join DataFrames                | `df.join(df2,"id","inner")`        |
| `union()`              | Stack DataFrames (same schema) | `df.union(df2)`                    |
| `unionByName()`        | Stack by column name           | `df.unionByName(df2)`              |

### 2. Column-Level Transformations (pyspark.sql.functions)

| Function              | Purpose             | Example                                |
| --------------------- | ------------------- | -------------------------------------- |
| `col()`               | Reference column    | `col("salary")`                        |
| `lit()`               | Constant column     | `lit(100)`                             |
| `when()`              | Conditional logic   | `when(col("x")>0,"Y")`                 |
| `concat()`            | Join strings        | `concat(col("a"),col("b"))`            |
| `concat_ws()`         | Join with separator | `concat_ws("-", "a","b")`              |
| `upper()` / `lower()` | Case change         | `upper(col("name"))`                   |
| `length()`            | String length       | `length(col("name"))`                  |
| `substring()`         | Extract substring   | `substring(col("x"),1,4)`              |
| `regexp_replace()`    | Regex replace       | `regexp_replace(col("x"),"[^0-9]","")` |
| `round()`             | Round value         | `round(col("x"),2)`                    |
| `ceil()` / `floor()`  | Math rounding       | `ceil(col("x"))`                       |
| `abs()`               | Absolute value      | `abs(col("x"))`                        |
| `date_format()`       | Format date         | `date_format(col("d"),"yyyy-MM")`      |
| `current_date()`      | Today’s date        | `current_date()`                       |
| `current_timestamp()` | Current time        | `current_timestamp()`                  |


# 4. Aggregation Functions (with groupBy())

| Function          | Purpose              | Example               |
| ----------------- | -------------------- | --------------------- |
| `sum()`           | Sum                  | `sum("salary")`       |
| `avg()`           | Average              | `avg("salary")`       |
| `min()`           | Minimum              | `min("salary")`       |
| `max()`           | Maximum              | `max("salary")`       |
| `count()`         | Count rows           | `count("*")`          |
| `countDistinct()` | Count unique         | `countDistinct("id")` |
| `collect_list()`  | Collect to array     | `collect_list("x")`   |
| `collect_set()`   | Collect unique array | `collect_set("x")`    |


5. Window Functions (Advanced Analytics)

| Function       | Purpose            | Example                |
| -------------- | ------------------ | ---------------------- |
| `row_number()` | Row ranking        | `row_number().over(w)` |
| `rank()`       | Rank with gaps     | `rank().over(w)`       |
| `dense_rank()` | Rank no gaps       | `dense_rank().over(w)` |
| `lag()`        | Previous row value | `lag("x",1).over(w)`   |
| `lead()`       | Next row value     | `lead("x",1).over(w)`  |
| `sum() over`   | Running total      | `sum("x").over(w)`     |
